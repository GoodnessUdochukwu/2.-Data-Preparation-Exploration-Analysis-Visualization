Change column names so that there are no names with spaces or weird special characters:

```python
smooth_column = []

for column in cars.columns:
  c = []
  for character in column:
    if character.isalpha() == True or character.isdigit() == True:
      c.append(character)
   
  smooth_column.append(''.join(c))

print(smooth_column)
cars.columns = smooth_column
```

Create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

- Create a plot that displays the evolution of the average CO2 emissions by cars over the years.

```python
emission_df = (
cars.groupby('Year')
    .agg(emission_aveg = ('CO2EmissionGramsMile', 'mean'))
    .reset_index()
)

emission_df
```

```python
emission_df.plot.line(x='Year', y='emission_aveg', figsize=(12, 10));
```

- Create a plot that clearly displays the evolution of the different types of fuel over the decades (only for the most common fuel types: "Diesel", "Gasoline or E85", "Premium", "Regular").

Note: you might have to transform the data shape before actually plotting. Use the column you created in the `intro-to-pandas` notebook with bins for each decade, and take a look at the [pivot_table](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html) function.

```python
most_common_fuel_types = cars[cars.FuelType.isin(["Diesel", "Gasoline or E85", "Premium", "Regular"])]
most_common_fuel_types
```

```python
most_common_fuel_types.Year_Period.value_counts()
```

```python
table = (
most_common_fuel_types.groupby(['Year_Period', 'FuelType'])
                      .agg(emission_aveg = ('CO2EmissionGramsMile', 'mean'))
                      .reset_index()
)
table                   
```

```python
most_common_fuel_types[(most_common_fuel_types.Year_Period == '90s') & (most_common_fuel_types.FuelType == 'Gasoline or E85')]
```

```python
table = table.dropna().reset_index(drop=True)
table
```

```python
table_norm = table.pivot(index='FuelType', columns='Year_Period', values='emission_aveg').fillna(0)
table_norm
```

```python
table_norm.plot.bar(figsize=(12, 10), rot=0);
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
table1 = table.pivot(columns='FuelType', index='Year_Period', values='emission_aveg').fillna(0)
table1
```

```python
ax = sns.barplot(data=table1, x=table1.index, y=table1.Diesel)
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('FuelType') # x label name
plt.ylabel('emission_aveg') # y label name
plt.show()
```

```python
ax = sns.catplot(x='FuelType', y='emission_aveg', hue='Year_Period', kind='bar', height = 8, aspect=1.2, data=table)
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('FuelType') # x label name
plt.ylabel('emission_aveg') # y label name
plt.show()
```

```python
ax = sns.catplot(x='Year_Period', y='emission_aveg', hue='FuelType', kind='bar', height = 8, aspect=1.2, data=table)
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('FuelType') # x label name
plt.ylabel('emission_aveg') # y label name
plt.show()
```
